```python
def bubblesort(a):
    n = len(a)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if a[j] > a[j + 1]:
                temp = a[j]
                a[j] = a[j + 1]
                a[j + 1] = temp

x = [23, 43, 25.78, 36, 75, 43, 96, 22, 75, 46, 6]
print('Before sorting', x)
bubblesort(x)
print('After sorting', x)
```

```python
def selectionsort(a):
    n = len(a)
    for i in range(n-1):
        min = i
        for j in range(i+1, n):
            if a[j] < a[min]:
                min = j
        a[i], a[min] = a[min], a[i]

x = [34, 46, 26, 37, 18, 49, 50, 48]
print("Before sorting:", x)
selectionsort(x)
print("After sorting:", x)
```

```python
def insertionsort(a):
    n = len(a)
    for i in range(1, n):
        key = a[i]
        j = i - 1
        while j >= 0 and key < a[j]:
            a[j + 1] = a[j]
            j = j - 1
        a[j + 1] = key

x = [34, 46, 26, 37, 18, 49, 50, 48]
print("Before sorting:", x)
insertionsort(x)
print("After sorting:", x)
```

```python
def binarysearch(arr, low, high, x):
  if high >= low:
    mid = (high + low) // 2
    if arr[mid] == x:
      return mid
    elif arr[mid] > x:
      return binarysearch(arr, low, mid - 1, x)
    else:
      return binarysearch(arr, mid + 1, high, x)
  else:
    return -1

arr = [2, 3, 4, 5, 10, 40]
x = 10
result = binarysearch(arr, 0, len(arr) - 1, x)
if result != -1:
  print("Element is present at index", str(result))
else:
  print("Element is not present in array")
```

```python
import json
import csv
import os
import re
import logging
import hashlib
from getpass import getpass
from datetime import datetime

# Configuration
DATABASE_FILE = "user_database.json"
ADMIN_CONFIG_FILE = "admin_config.json"
LOG_FILE = "user_management.log"
CSV_EXPORT_FILE = "user_export.csv"

# Initialize logging
logging.basicConfig(
    filename=LOG_FILE,
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

class UserManager:
    def __init__(self):
        self.user_database = self.load_users()
        self.admin_config = self.load_admin_config()
        self.current_user = None
        self.admin_authenticated = False

    #region Core Functionality
    def load_users(self):
        if os.path.exists(DATABASE_FILE):
            with open(DATABASE_FILE, "r") as file:
                return json.load(file)
        return []

    def save_users(self):
        with open(DATABASE_FILE, "w") as file:
            json.dump(self.user_database, file, indent=4)

    def load_admin_config(self):
        if os.path.exists(ADMIN_CONFIG_FILE):
            with open(ADMIN_CONFIG_FILE, "r") as file:
                return json.load(file)
        return {"password_hash": "", "hint": "Default hint: Contact system administrator"}

    def save_admin_config(self):
        with open(ADMIN_CONFIG_FILE, "w") as file:
            json.dump(self.admin_config, file, indent=4)
    #endregion

    #region Authentication
    def authenticate_admin(self):
        if self.admin_authenticated:
            return True

        print("\nAdmin Login")
        password = getpass("Enter admin password: ")
        input_hash = hashlib.sha256(password.encode()).hexdigest()

        if input_hash == self.admin_config["password_hash"]:
            self.admin_authenticated = True
            logging.info("Admin authenticated successfully")
            return True

        print(f"Hint: {self.admin_config['hint']}")
        logging.warning("Failed admin authentication attempt")
        return False

    def authenticate_user(self):
        print("\nUser Login")
        reg_no = input("Registration number: ").strip()
        password = getpass("Password: ")

        user = next((u for u in self.user_database if u["reg_no"] == reg_no), None)
        if not user:
            print("User not found!")
            return False

        input_hash = hashlib.sha256(password.encode()).hexdigest()
        if user["password_hash"] == input_hash:
            self.current_user = user
            logging.info(f"User logged in: {reg_no}")
            return True

        print("Invalid credentials!")
        return False
    #endregion

    #region User Operations
    def add_user(self):
        if not self.authenticate_admin():
            return

        user = {
            "name": input("Name: ").strip(),
            "reg_no": "",
            "address": input("Address: ").strip(),
            "phone_no": "",
            "password_hash": ""
        }

        # Registration number validation
        while True:
            reg_no = input("Registration number: ").strip()
            if self.is_unique_reg_no(reg_no):
                user["reg_no"] = reg_no
                break
            print("Registration number already exists!")

        # Phone validation
        while True:
            phone = input("Phone number: ").strip()
            if re.match(r"^\d{10}$", phone):
                user["phone_no"] = phone
                break
            print("Invalid phone number (must be 10 digits)!")

        # Password setup
        while True:
            password = getpass("Set password: ")
            if len(password) >= 8:
                user["password_hash"] = hashlib.sha256(password.encode()).hexdigest()
                break
            print("Password must be at least 8 characters!")

        self.user_database.append(user)
        self.save_users()
        logging.info(f"User added: {reg_no}")
        print("User added successfully!\n")

    def keyword_search(self, keyword):
        keyword = keyword.lower()
        return [user for user in self.user_database if
                keyword in user["name"].lower() or
                keyword in user["reg_no"].lower() or
                keyword in user["address"].lower() or
                keyword in user["phone_no"].lower()]

    def advanced_search(self, criteria):
        return [user for user in self.user_database if all(
            str(user.get(k, "")).lower().find(v.lower()) != -1
            for k, v in criteria.items() if v
        )]

    def edit_user(self, reg_no):
        user = next((u for u in self.user_database if u["reg_no"] == reg_no), None)
        if not user:
            print("User not found!")
            return

        print("\nEditable Fields:")
        fields = {
            1: ("name", "Name"),
            2: ("reg_no", "Registration Number"),
            3: ("address", "Address"),
            4: ("phone_no", "Phone Number"),
            5: ("password", "Password")
        }

        for k, v in fields.items():
            print(f"{k}. {v[1]}")

        try:
            choice = int(input("\nSelect field to edit: "))
            field = fields[choice][0]
        except (ValueError, KeyError):
            print("Invalid selection!")
            return

        if field == "password":
            new_pass = getpass("New password: ")
            user["password_hash"] = hashlib.sha256(new_pass.encode()).hexdigest()
        else:
            new_value = input(f"New {fields[choice][1]}: ").strip()
            if field == "phone_no" and not re.match(r"^\d{10}$", new_value):
                print("Invalid phone number!")
                return
            user[field] = new_value

        self.save_users()
        logging.info(f"User updated: {reg_no}")
        print("Update successful!\n")

    def delete_user(self, reg_no):
        user = next((u for u in self.user_database if u["reg_no"] == reg_no), None)
        if not user:
            print("User not found!")
            return

        confirm = input(f"Delete {user['name']}? (y/n): ").lower()
        if confirm == "y":
            self.user_database.remove(user)
            self.save_users()
            logging.info(f"User deleted: {reg_no}")
            print("User deleted successfully!\n")
    #endregion

    #region Admin Management
    def change_admin_password(self):
        if not self.authenticate_admin():
            return

        new_pass = getpass("New admin password: ")
        confirm = getpass("Confirm password: ")

        if new_pass != confirm:
            print("Passwords don't match!")
            return

        self.admin_config["password_hash"] = hashlib.sha256(new_pass.encode()).hexdigest()
        self.admin_config["hint"] = input("Password hint: ").strip()
        self.save_admin_config()
        logging.info("Admin password changed")
        print("Password updated successfully!\n")
    #endregion

    #region File Operations
    def export_csv(self):
        with open(CSV_EXPORT_FILE, "w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["name", "reg_no", "address", "phone_no", "password_hash"])
            writer.writeheader()
            writer.writerows(self.user_database)
        logging.info("Database exported to CSV")
        print(f"Data exported to {CSV_EXPORT_FILE}")

    def import_csv(self):
        try:
            with open(CSV_EXPORT_FILE, "r") as file:
                reader = csv.DictReader(file)
                imported = 0
                for row in reader:
                    if all(f in row for f in ["name", "reg_no", "password_hash"]) and self.is_unique_reg_no(row["reg_no"]):
                        self.user_database.append(row)
                        imported += 1
                self.save_users()
                logging.info(f"Imported {imported} users from CSV")
                print(f"Successfully imported {imported} users!")
        except FileNotFoundError:
            print("CSV file not found!")
    #endregion

    #region Utilities
    def sort_users(self, field):
        valid_fields = ["name", "reg_no", "address", "phone_no"]
        if field not in valid_fields:
            return None
        return sorted(self.user_database, key=lambda x: x[field].lower())

    def is_unique_reg_no(self, reg_no):
        return all(user["reg_no"] != reg_no for user in self.user_database)

    def get_user_details(self, user):
        return f"""
        Name: {user['name']}
        Registration Number: {user['reg_no']}
        Address: {user['address']}
        Phone Number: {user['phone_no']}
        """
    #endregion

#region Interface Functions
def admin_menu(manager):
    while True:
        print("\nADMIN PANEL")
        print("1. Add User")
        print("2. Quick User Search")
        print("3. Advanced User Search")
        print("4. Edit User")
        print("5. Delete User")
        print("6. List All Users")
        print("7. Export to CSV")
        print("8. Import from CSV")
        print("9. Change Admin Password")
        print("10. Logout")

        choice = input("\nSelect option: ").strip()

        if choice == "1":
            manager.add_user()

        elif choice == "2":
            keyword = input("\nSearch keyword: ").strip()
            results = manager.keyword_search(keyword)
            print(f"\nFound {len(results)} users:")
            for user in results:
                print(manager.get_user_details(user))

        elif choice == "3":
            print("\nAdvanced Search (leave fields blank)")
            criteria = {
                "name": input("Name contains: "),
                "reg_no": input("Registration number contains: "),
                "address": input("Address contains: "),
                "phone_no": input("Phone contains: ")
            }
            results = manager.advanced_search(criteria)
            print(f"\nFound {len(results)} users:")
            for user in results:
                print(manager.get_user_details(user))

        elif choice == "4":
            reg_no = input("Registration number to edit: ").strip()
            manager.edit_user(reg_no)

        elif choice == "5":
            reg_no = input("Registration number to delete: ").strip()
            manager.delete_user(reg_no)

        elif choice == "6":
            field = input("Sort by (name/reg_no/address/phone_no): ").lower()
            sorted_users = manager.sort_users(field)
            for user in sorted_users or []:
                print(manager.get_user_details(user))

        elif choice == "7":
            manager.export_csv()

        elif choice == "8":
            manager.import_csv()

        elif choice == "9":
            manager.change_admin_password()

        elif choice == "10":
            manager.admin_authenticated = False
            break

        else:
            print("Invalid option!")

def user_menu(manager):
    print("\nUSER PANEL")
    print(manager.get_user_details(manager.current_user))
    input("Press Enter to logout...")

def main():
    manager = UserManager()

    while True:
        print("\nMAIN MENU")
        print("1. Admin Login")
        print("2. User Login")
        print("3. Exit")

        choice = input("Select option: ").strip()

        if choice == "1":
            if manager.authenticate_admin():
                admin_menu(manager)

        elif choice == "2":
            if manager.authenticate_user():
                user_menu(manager)
                manager.current_user = None

        elif choice == "3":
            print("Exiting system...")
            break

        else:
            print("Invalid option!")

if __name__ == "__main__":
    # First-time setup check
    if not os.path.exists(ADMIN_CONFIG_FILE):
        print("Initial setup required!")
        manager = UserManager()
        manager.change_admin_password()

    main()
#endregion
```
